```python
#To detect the anomalies
plt.figure(figsize=(15,8))

plt.subplot(1,3,1)
sns.boxplot( y=data["math score"] )

plt.subplot(1,3,2)
sns.boxplot( y=data["reading score"] )

plt.subplot(1,3,3)
sns.boxplot( y=data["writing score"] )
```

1.The balck spots in the above box plot are the anomalies.
IQR is the inter quartile range.

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#to remove the anomalies
(data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))
```

```python
filtered_data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
filtered_data.shape
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,3,1)
sns.boxplot( y=filtered_data["math score"] )

plt.subplot(1,3,2)
sns.boxplot( y=filtered_data["reading score"] )

plt.subplot(1,3,3)
sns.boxplot( y=filtered_data["writing score"] )
```
